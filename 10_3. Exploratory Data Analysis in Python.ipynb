## Addressing missing data


```python
salaries=pd.read_csv('G:\\datacamp\\10_3. Exploratory Data Analysis\\ds_salaries_clean.csv')
salaries.head()
```

```python
salaries.info()
```

```python
print(salaries.isna().sum())
threshold=len(salaries)*.05  ## we can delete missing values upto 30
threshold ## there is no missing value in the data set
```

```python
salaries.columns
```

WE CAN HANDLE MISSING VALUES  BY MAINLY THREE WAYS:
1. drop 2.Impute mean, median or mode, 3. Impute by sub groups

```python
## drop
cols_to_drop=salaries.columns[salaries.isna().sum()<threshold] # select the columns that can be drop. 
                                                               #though it has no missing values
print(cols_to_drop)
salaries.dropna(subset=cols_to_drop, inplace=True) # nplace =true so that df is updated
```

```python
## impute
cols_with_more_missing_values=salaries.columns[salaries.isna().sum()>0]
for col in cols_with_more_missing_values[:-1]: # without taking the last column/Salary_usd
    salaries[col].fillna(salaries[col].mode()[0])
    
```

In Pandas, df['col'] is used to extract a column named 'col' from a DataFrame df.

.mode() is a method that can be applied to a Pandas Series (i.e., a column of a DataFrame) to calculate the mode of the column, which is the most frequently occurring value. If there are multiple values that occur with the same highest frequency, .mode() will return all of them.

The [0] at the end of df['col'].mode()[0] is used to extract the first (i.e., most frequent) value from the mode array returned by .mode().

So df['col'].mode()[0] returns the most frequent value of the 'col' column in the DataFrame df.


```python
## imputing by sub groups
salaries_dict=salaries.groupby('Experience')['Salary_USD'].median().to_dict()
print(salaries_dict)
salaries['Salary_USD']=salaries['Salary_USD'].fillna(salaries['Experience'].map(salaries_dict)) ## using map to fill the missing salaries with median
```

#### Removing outliers
While removing outliers isn't always the way to go, for your analysis, you've decided that you will only include flights where the "Price" is not an outlier.

Therefore, you need to find the upper threshold and then use it to remove values above this from the planes DataFrame.

```python
# Find the 75th and 25th percentiles
price_seventy_fifth = planes["Price"].quantile(0.75)
price_twenty_fifth = planes["Price"].quantile(0.25)

# Calculate iqr
prices_iqr = price_seventy_fifth - price_twenty_fifth

# Calculate the thresholds
upper = price_seventy_fifth + (1.5 * prices_iqr)
lower = price_twenty_fifth - (1.5 * prices_iqr)

# Subset the data
planes = planes[(planes["Price"] > lower) & (planes["Price"] < upper)]

print(planes["Price"].describe())
```

```python
sns.histplot(data=planes,x='Price')
plt.show()
```

### Kernal Density Plot(KDE)
because the education levels are stacked on top of each other, the relationship between marriage duration and male education level isn't super clear.

<b> KDE </b> <br>
Seaborn's Kernel Density Estimate or KDE plots address this issue. Similar to histograms, KDEs allow us to visualize distributions. KDEs are considered more interpretable, though, especially when multiple distributions are shown as they are here. Notice that the location of the peak marriage duration for each level of the male partner's education is more identifiable in this KDE plot than it was in the histogram. However, due to the smoothing algorithm used in KDE plots, the curve can include values that don't make sense, so it's important to set good smoothing parameters.

```python
sns.kdeplot(data=divorce_,x='marriage_duration',hue='education_man')
plt.show()
```

but the graphs are started before zero. that is not possible. we can minimize it by using cut. the curve will be limited to values between the minimum and maximum x values.<br>The cut parameter in sns.kdeplot() controls the extent of the density estimate on the x-axis. Specifically, it sets the number of bandwidths beyond the extreme data points where the density estimate is computed. The default value for cut is 3, which means that the density estimate is computed up to three bandwidths beyond the extreme data points.<br>

If you increase the value of cut, the density estimate will be computed farther beyond the range of the data, resulting in a smoother curve that extends further out on the x-axis. Conversely, if you decrease the value of cut, the density estimate will be computed for a smaller range of values, resulting in a less smooth curve that stops closer to the range of the data.<br>

For example, if you set cut=1, the density estimate will be computed up to one bandwidth beyond the extreme data points, resulting in a curve that is less smooth and stops closer to the range of the data. Similarly, if you set cut=2, the density estimate will be computed up to two bandwidths beyond the extreme data points, resulting in a slightly smoother curve that extends a bit further out on the x-axis.

```python
sns.kdeplot(data=divorce_,x='marriage_duration',hue='education_man',cut=0)
plt.show()
```

```python
# cumulative
sns.kdeplot(data=divorce_,x='marriage_duration',hue='education_man',cut=0,cumulative=True)
plt.show()
```
