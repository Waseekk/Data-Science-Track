#  2//Visualizing Two Quantitative Variables
##  2.1 Introduction to relational plots and subplots
"relplot()" stands for "relational plot" and enables you to visualize the relationship between two quantitative variables using either scatter plots or line plots. Using "relplot()" gives us a big advantage: the ability to create subplots in a single figure. Because of this advantage, we'll be using "relplot()" instead of "scatterplot()" for the rest of the course.

```python
# Change to make subplots based on study time
sns.relplot(x="absences", y="G3",     
            data=student_data,
            kind="scatter")

# Show plot
plt.show()
```

THE BEST THING  about this replot() is we can create one scatter plot for each level by definning row and col,umn

```python
# Change this scatter plot to arrange the plots in rows instead of columns
sns.relplot(x="absences", y="G3", 
            data=student_data,
            kind="scatter", 
            col='study_time')
           # row="study_time")

# Show plot
plt.show()
```

col_order to control the order of yes and no. see we can use col and row. even row_order also.

```python
# Adjust further to add subplots based on family support
sns.relplot(x="G1", y="G3", 
            data=student_data,
            kind="scatter", 
            col="schoolsup",
            col_order=["yes", "no"])

# Show plot
plt.show()
```

```python
# Adjust further to add subplots based on family support
sns.relplot(x="G1", y="G3", 
            data=student_data,
            kind="scatter", 
            col="schoolsup",
            col_order=["yes", "no"],row='famsup',row_order=['yes','no'])

# Show plot
plt.show()
```

## 3.2 Box plot
A box plot shows the distribution of quantitative data. The colored box represents the 25th to 75th percentile, and the line in the middle of the box represents the median. The whiskers give a sense of the spread of the distribution, and the floating points represent outliers. Box plots are commonly used as a way to compare the distribution of a quantitative variable across different groups of a categorical variable.

```python
# Specify the category ordering
study_time_order = ["<2 hours", "2 to 5 hours", 
                    "5 to 10 hours", ">10 hours"]

# Create a box plot and set the order of the categories
sns.catplot(x='study_time',y='G3',data=student_data,kind='box',order=study_time_order)
```

see the powewr of hue and sym. sym='' means all the outliers is omitted. and hue is for coloring.

```python
# Create a box plot with subgroups and omit the outliers

sns.catplot(x='internet',y='G3',data=student_data,kind='box',sym="",hue='location') ## check it without hue




# Show plot
plt.show()
```

By default, the whiskers extend to 1 point 5 times the interquartile range, or "IQR". The IQR is the 25th to the 75th percentile of a distribution of data. If you want to change the way the whiskers in your box plot are defined, you can do this using the "whis" parameter. ALso you may just want to draw the whiskers at the min and max values. 

```python
# Set the whiskers to 0.5 * IQR
sns.catplot(x="romantic", y="G3",
            data=student_data,
            kind="box",whis=.5)   ##whis=[0,100] means raw the whis at mina nd max value

# Show plot
plt.show()
```

## 3.3 point plot
### Point plots vs. line plots
Both line plots and point plots show the mean of a quantitative variable and 95% confidence intervals for the mean. However, there is a key difference. Line plots are relational plots, so both the x- and y-axis are quantitative variables. In a point plot, one axis - usually the x-axis - is a categorical variable, making it a categorical plot.


#### Point plot vs box plot
In the point plot, it's easier to compare the heights of the subgroup points when they're stacked above each other. In the point plot, it's also easier to look at the differences in slope between the categories than it is to compare the heights of the bars between them.


```python
sns.catplot(x='famrel',y='absences',data=student_data,kind='point')  ##family rltn with the absences
            
# Show plot
plt.show()
```

```python
# Add caps to the confidence interval
sns.catplot(x="famrel", y="absences",
			data=student_data,
            kind="point",capsize=.2)
        
# Show plot
plt.show()
```

Here, we can compare between different values on x axis. but if we only need the result for a certain category. then we dont need the join. right???


```python
# Remove the lines joining the points
sns.catplot(x="famrel", y="absences",
			data=student_data,
            kind="point",
            capsize=0.2,join=False)
            
# Show plot
plt.show()
```

## 4.2//Adding titles and labels: Part 1
Seaborn's plot functions create two different types of objects: **FacetGrids and AxesSubplots.** To figure out which type of object you're working with, first assign the plot output to a variable.A FacetGrid consists of one or more AxesSubplots, which is how it supports subplots.

Recall that "relplot()" and "catplot()" both support making subplots. This means that they are creating FacetGrid objects. In contrast, single-type plot functions like "scatterplot()" and "countplot()" return a single AxesSubplot object.

```python


# Create scatter plot
g = sns.relplot(x="weight", 
                y="horsepower", 
                data=mpg,
                kind="scatter")

# Add a title "Car Weight vs. Horsepower"
print(type(g))
g.fig.suptitle('Car Weight vs. Horsepower')
# Show plot
plt.show()
```

```python
# Create point plot
sns.catplot(x="origin", 
            y="acceleration", 
            data=mpg, 
            kind="point", 
            join=False, 
            capsize=0.1)

# Rotate x-tick labels
plt.xticks(rotation=90)  ##for rotating we can call the plt function of matplotlib

# Show plot
plt.show()
```
